```
import tensorflow as tf
import os
from   random import shuffle
import shutil
```

```
def get_num_files(path):
  return len(os.listdir(path))
```

```
def get_each_file_count(path):
  for subdir in os.listdir(path):
    subdirPath=os.path.join(path,subdir)
    length=get_num_files(subdirPath)
    print("Number of images in {0} is {1}".format(subdir,length))
```

```
get_each_file_count("/content/malimg_paper_dataset_imgs")
```

```
def create_test_set(path,count=20):
  test_parent="test"
  try:
    os.mkdir("test")
  except FileExistsError:
    shutil.rmtree(test_parent,ignore_errors=True)
    os.mkdir("test")
  for subdir in os.listdir(path):
    files_in_subdir_20=os.listdir(os.path.join(path,subdir))[:20]
    os.mkdir(os.path.join(test_parent,subdir))
    for filename in files_in_subdir_20:
      #print(os.path.join(test_parent,subdir,filename))
      os.rename(os.path.join(path,subdir,filename),os.path.join(test_parent,subdir,filename))
```

```
create_test_set("/content/malimg_paper_dataset_imgs")
```

```
get_each_file_count("/content/test")
```

```
aug       = tf.keras.preprocessing.image.ImageDataGenerator(validation_split=0.3, rescale=1/255.)
train_gen = aug.flow_from_directory("/content/malimg_paper_dataset_imgs",
                                    subset="training", class_mode="sparse", color_mode="grayscale",
                                    target_size=(32,32))
val_gen   = aug.flow_from_directory("/content/malimg_paper_dataset_imgs",
                                    subset="validation", class_mode="sparse",color_mode="grayscale",
                                    target_size=(32,32))
test_gen  = aug.flow_from_directory("/content/test",
                                    class_mode="sparse",color_mode="grayscale",
                                    target_size=(32,32))
```

```
train_gen.class_indices
```

```
def build_model():
  input_layer=tf.keras.layers.Input((32,32,1))
  x=tf.keras.layers.Conv2D(32,(2,2),activation="relu",name="mal_conv_0")(input_layer)
  x=tf.keras.layers.MaxPool2D()(x)
  x=tf.keras.layers.Conv2D(16,(2,2),activation="relu",name="mal_conv_1")(x)
  x=tf.keras.layers.MaxPool2D()(x)
  x=tf.keras.layers.Flatten(name="mal_flat_0")(x)
  x=tf.keras.layers.Dense(128,activation="relu", name="mal_dense_0")(x)
  x=tf.keras.layers.Dropout(0.2, name="mal_dropout_0")(x)
  x=tf.keras.layers.Dense(64,activation="relu", name="mal_dense_1")(x)
  x=tf.keras.layers.Dropout(0.2, name="mal_dropout_1")(x)
  x=tf.keras.layers.Dense(25, activation="softmax", name="mal_dense_2")(x)
  malnet=tf.keras.models.Model(inputs=input_layer, outputs=x)
  malnet.compile(loss="sparse_categorical_crossentropy", metrics=["accuracy"], optimizer="adam")
  return malnet
```

```
malnet=build_model()
```

```
malnet.summary()
```

```
!mkdir models
```

```
callback = tf.keras.callbacks.ModelCheckpoint("models/weights.{epoch:02d}-{val_loss:.2f}.hdf5")
```

```
history=malnet.fit_generator(train_gen, epochs=20, callbacks=[callback], validation_data=val_gen)
```

```
model = tf.keras.models.load_model("/content/models/weights.18-0.39.hdf5")
```

```
model.evaluate_generator(test_gen)
```
